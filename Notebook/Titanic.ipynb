```python
Titanic_copy = Titanic_copy.drop(columns=['Ticket', 'Cabin'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,3))
sns.histplot(Titanic_copy['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
Titanic_copy['Age'].describe()
```

```python
Titanic_copy['Age'] = Titanic_copy['Age'].fillna(Titanic_copy['Age'].median())
```

```python
Titanic_copy['Age'] = Titanic_copy['Age'].astype(int)
```

```python
Titanic_copy.head(5)
```

```python
Titanic_copy['Title'] = Titanic_copy['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
Titanic_copy['Title'].value_counts()
```

```python
Titanic_copy['Title'].isnull().any()
```

```python
Titanic_copy['Title'] = Titanic_copy['Title'].replace({'Mlle': 'Miss',
                                                      'Ms': 'Miss',
                                                     'Mme': 'Mrs'})
```

```python
Titanic_copy['Title'] = Titanic_copy['Title'].replace(
    ['Lady', 'Countess','Capt', 'Col',
     'Don', 'Dr', 'Major', 'Rev', 'Sir',
     'Jonkheer', 'Dona'], 'Rare')
```

```python
Titanic_copy.head()
```

```python
Titanic_copy['Embarked'].value_counts()
```

```python
Titanic_copy['Embarked'].mode()
```

```python
Titanic_copy['Embarked'] = Titanic_copy['Embarked'].fillna(Titanic_copy['Embarked'].mode()[0])
```

### Exploratory Data Analysis

#### Survived vs Non-survived

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.countplot(x='Survived', data=Titanic_copy)
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

#### Gender distribution of passengers 

```python
ax = sns.countplot(x='Sex', data=Titanic_copy)
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

#### Age distribution of passengers

```python
ax = sns.histplot(Titanic_copy['Age'], kde=True)
for c in ax.containers:
    ax.bar_label(c)
plt.show()
```

#### Survival rate by sex

```python
ax = sns.barplot(x='Sex', y='Survived', data=Titanic_copy, errorbar=None)

survivors = Titanic_copy.query("Survived == 1")['Sex'].value_counts()

for container in ax.containers:
    ax.bar_label(container, labels=[f"{survivors[sex]} survived" for sex in survivors.index])

plt.show()
```

#### Survival rate by Pclass

```python
ax = sns.barplot(x='Pclass', y='Survived', data=Titanic_copy, errorbar=None)
for c in ax.containers:
    ax.bar_label(c, fmt='%.2f')
plt.show()
```

#### Fare by Survived

```python
sns.boxplot(x='Survived', y='Fare', data=Titanic_copy)
```

#### Survival rate by embarked

```python
ax = sns.barplot(x='Embarked', y='Survived', data=Titanic_copy, errorbar=None)
for c in ax.containers:
    ax.bar_label(c, fmt='%.2f')
plt.show()
```

#### Family size vs Survival

```python
Titanic_copy['Family_size'] = Titanic_copy['SibSp'] + Titanic_copy['Parch'] + 1
Titanic_copy.groupby('Family_size')['Survived'].mean()
```

#### Survival rate by Pclass and sex(Hue)

```python
ax = sns.barplot(x='Pclass', y='Survived', hue='Sex', data=Titanic_copy, errorbar=None)
for c in ax.containers:
    ax.bar_label(c, fmt='%.2f')
plt.show()
```

#### Survival rate by title

```python
ax = sns.barplot(x='Title', y='Survived', data=Titanic_copy, errorbar=None)
for c in ax.containers:
    ax.bar_label(c, fmt='%.2f')
plt.show()
```
